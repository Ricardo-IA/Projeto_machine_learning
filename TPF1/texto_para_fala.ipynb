```python
from gtts import gTTS

text_to_say = "Se é para o bem de todos e felicidade geral da nação, digo ao povo que fico!."

language = "pt"

gtts_object = gTTS(text = text_to_say, 
                  lang = language,
                  slow = False)
gtts_object.save("/content/gtts.wav")
```

```python
from IPython.display import Audio

Audio("/content/gtts.wav")
```

```python
from gtts import gTTS
french_text = "Si c'est pour le bien de tous et le bonheur général de la nation, je dis au peuple que je reste !"

french_language = "fr"

french_gtts_object = gTTS(text = french_text,
                          lang = french_language,
                          slow = True)

french_gtts_object.save("/content/french.wav")

Audio("/content/french.wav")
```
